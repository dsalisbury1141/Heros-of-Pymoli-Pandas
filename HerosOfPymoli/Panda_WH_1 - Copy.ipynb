## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

# Age Buckets
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
buckets = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

purchase_df=purchase_df.drop_duplicates(subset='SN', keep="first")
numOfRows = len(purchase_df.index)
purchase_df["Age Demographics"] = pd.cut(purchase_df["Age"],bins=age_bins, labels=buckets)

AgeTotals=purchase_df["Age Demographics"].value_counts(sort=False)
PPlayers= AgeTotals/numOfRows*100
purchase_df.style.format({'PPlayers':'{:,.2f}%'})


AgeSum=pd.DataFrame({"Age Demographics":AgeTotals,"Percentage of Players":PPlayers})
#PPlayers["Percentage of Players"] = PPlayers["Percentage of Players"].map("{:,.2f}%".format)
AgeSum

```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

# Age Buckets
age_bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 999]
buckets = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

#Analyzing data
#numOfRows = len(purchase_df.index) 
purchase_df["Purchase Analysis(Age)"] = pd.cut(purchase_df["Age"],bins=age_bins, labels=buckets)
AgeTotal=purchase_df["Purchase Analysis(Age)"].value_counts(sort=False)


Agegroup = purchase_df.groupby("Purchase Analysis(Age)")
Ageotalprice = Agegroup['Price'].sum()
Ageavgperson = Ageotalprice/Agegroup['Item Name'].nunique()
Ageavgprice = Ageotalprice/Agegroup["SN"].nunique()

# formating
Ageotalprice = Ageotalprice.map('${:.2f}'.format)
Ageavgprice = Ageavgprice.map('${:.2f}'.format)
Ageavgperson = Ageavgperson.map('${:.2f}'.format)

#Printing data
PurchaseSum=pd.DataFrame({'Purchase Count':AgeTotal
                          ,"Average Purchase Price":Ageavgperson
                          , "Average Purchase Value":Ageotalprice
                         ,"Average Total Purchase Per Person":Ageavgprice})
PurchaseSum
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Total purchases by screen Name
purchases = purchase_df.groupby("SN")["Purchase ID"].count()


#Average price per Screen Name
Taverage = round(purchase_df.groupby("SN")["Price"].mean(),2)


Tprice = purchase_df.groupby("SN")["Price"].sum()


#Taverage = Tprice/purchases
#Taverage

TotalPurchase=Taverage/purchases


#Print data
TopSpendersSummary=pd.DataFrame({"Purchase Count":purchases
                                 ,"Average Purchase Price":Taverage
                                 , "Total Purchase Value":TotalPurchase})

#TopSpendersSummary

sorted_summary = TopSpendersSummary.sort_values(["Purchase Count"], ascending=True)
sorted_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
 
#Popular list
PopularName=purchase_df[["Item ID", "Item Name", "Price"]]

#Popular Items
purchases = PopularName.groupby(["Item ID", "Item Name"])["Price"].count()


#Popular price
PopularPrice = round(purchase_df.groupby(["Item ID", "Item Name"])["Price"].mean(),2)


#popular total
Tprice=purchase_df.groupby(["Item ID", "Item Name"])["Price"].sum()




#Print data
PopularSummary=pd.DataFrame({"Purchase Count":purchases
              , "Item Price":PopularPrice
              , "Total Purchase Value":Tprice,})


SortedPopularSummary = PopularSummary.sort_values("Purchase Count", ascending=False)
SortedPopularSummary.head()
```
